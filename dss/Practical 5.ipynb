```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simple Line Plot
x = [2,4,6,8,10]
y = [14,87,65,32,19]

plt.plot(x, y, color='green', marker='o', linewidth=3, linestyle='--', markersize=10)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Simple Line Graph")
```

```python
x = [2,4,6,8,10]
y = [15,64,63,16,57]

x1 = [2,4,6,8,10]
y1 = [57,45,68,31,35]

plt.plot(x, y, color='green', marker='o', linewidth=3)
plt.plot(x1, y1, color='coral', marker='^', linewidth=3)
plt.grid('true')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Multiple Lines Graph")
```

```python
# Bar Graph
x = [2,4,6,8,10]
y = [14,87,65,32,19]

plt.bar(x, y, color='purple', width=1.5)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Simple Bar Graph")
```

```python
x = [2,4,6,8,10]
y = [15,64,63,16,57]

plt.barh(x, y, color='orange', height=1.5)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Horizontal Bar Graphs")
```
