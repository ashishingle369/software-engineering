# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the datasets
df = pd.read_csv(r"D:\3N44\Data Sets\titanic.csv")
df.head()
```

```python
df.info()
```

```python
# Handling Missing Values
df.isnull().sum()
```

```python
# Fill missing Age values with the median age
df['Age'].fillna(df['Age'].median(), inplace= True)
df.isnull().sum()
```

```python
# Fill missing Embarked values with the most frequent port
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace= True)
df.isnull().sum()
```

```python
# Survival by Gender
sns.countplot(x='Gender',hue='Survived',data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
# Survival by Pclass
sns.countplot(x='Pclass',hue='Survived',data=df)
plt.title("Survival by Pclass")
plt.show()
```

```python
# Distribution of Age
sns.histplot(df['Age'],kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
# Age Distribution of Gender and Pclass
sns.violinplot(x='Pclass',y='Age',hue='Gender',data=df)
plt.title('Age Distribution of Gender and Pclass')
plt.show()
```

```python
# Correlation heatmap of numerical features
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
